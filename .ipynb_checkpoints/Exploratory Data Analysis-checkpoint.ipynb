## Section 1: Pre-processing, Feature Analysis and Visualizations


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import nltk, os, re, string, time
from nltk.util import ngrams 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from textblob import Word
from collections import Counter

%matplotlib inline
warnings.filterwarnings('ignore')
```

### Read dataframe

```python
# read data
df = pd.read_json('News_Category_Dataset_v2.json', lines=True)
```

```python
df.head()
```

### Preprocess dataset

```python
def replace_bad_chars(string):
    return re.sub(r'[^a-zA-Z0-9\s]', ' ', string)

def preprocess_data(df):
    """
    Applies simple NLP preprocessing techniques on input dataframe.

    Args:
        df: input dataframe
    """
    ## aggregate headline and short_descriptions cols
    df['text'] = df.headline + " " + df.short_description

    ## get rid of non-alphanumeric chars
    df['text'] = df['text'].apply(replace_bad_chars)

    ## convert to lowercase
    df['text'] = df['text'].str.lower() 

    #3remove stop words
    stop_words = set(stopwords.words('english'))
    df['text'] = df['text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop_words))

    ## lemmatize text
    df['text'] = df['text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))

    ## tokenize text
    df['tokenized_text'] = df['text'].apply(word_tokenize) 

    ## get length of text and get rid of bad text based on length threshold
    ## threshold = 2 standard deviations below mean length
    df['text_length'] = df['tokenized_text'].apply(len)
    threshold = df['text_length'].mean() - 2*(df['text_length'].std())
    df = df[df.text_length >= threshold]

    return df
```

```python
preproc_df = preprocess_data(df)
preproc_df.head(3)
```

```python
ext = ".com"
preproc_df['source_link'] = preproc_df.link.str.partition(ext)[0] + ext
preproc_df.head(3)
```

```python
preproc_df['source_link'].value_counts(normalize=True)
```

### Category feature analysis

```python
preproc_df['category'].value_counts(normalize=True)
```

#### After breaking the dataset down by category labels, there are a few things that stand out:
    
1. If we were trying to predict a category label (given a headline and short description), it's immediately apparent that this is a **class-imbalance problem**. Politics news makes up over 16% of the dataset, while Tech news makes up only about 1%. We could tackle a problem like this with different sampling techniques (e.g. oversampling, undersampling, SMOTE).
2. Some categories are mislabeled. For example 'THE WORLDPOST' and 'WORLDPOST' should be grouped together. 
3. Some categories are subsets of others (e.g. 'Arts' and 'Arts & Culture'). Thus, in some sort of multi-classification problem, we might be more precise with the category labeling and group these together.

```python
mapping = {'THE WORLDPOST' : 'WORLDPOST', 'CULTURE & ARTS' : 'ARTS & CULTURE'}
preproc_df['category'] = preproc_df['category'].replace(mapping)
```

#### We can visualize the categories with a simple bar chart

```python
figure, ax = plt.subplots(1, 1, figsize = (15, 5))

ax = sns.countplot(preproc_df['category'])
ax.set_title("Category frequency")
ax.set_xlabel("Category")
ax.set_xticklabels([x for x in ax.get_xticklabels()], rotation=75)
plt.show()
```

### Author feature analysis

<p> One initial intuition of mine when I saw the features in the dataset was to check if certain authors were responsible for publishing a disproportionate # of news articles. It's also likely that certain authors write articles for specific categories, so a correlation analysis of those two features would be interesting. </p>

```python
preproc_df.authors.value_counts(normalize = True).tail()
```

<p> **Note: **The last few author labels are faulty, so we can do some string manipulation to clean those labels up. </p>

```python
preproc_df['authors'] = preproc_df['authors'].str.partition(",")
```

```python
preproc_df['authors'].value_counts(normalize = True).tail()
```

```python
num_authors = len(preproc_df['authors'].unique())
num_categories = len(preproc_df['category'].unique())
grouped_df = preproc_df.groupby('authors')['category'].nunique().reset_index(name='unique_category_count')
grouped_df = grouped_df.sort_values(by = ['unique_category_count'], ascending = False)
grouped_df['percent_category_coverage'] = grouped_df['unique_category_count'] / num_categories
```

```python
grouped_df.head()
```

```python
grouped_df.tail()
```

```python
single_category_authors = grouped_df.loc[grouped_df['unique_category_count'] == 1]
sing_categ_authors_count = len(single_category_authors['authors'].unique())
percent_sing_authors = (sing_categ_authors_count / num_authors) * 100
print("-----------------")
print("{0}% of authors published articles for one specific category! ".format(round(percent_sing_authors)))
print("-----------------")
```

<p> This is an interesting result. The top row of the *grouped_df* with a null value under the authors field are for news articles that had no listed author. We see that articles with no author listed cover all categories in the dataset (100% category coverage). We also see that certain authors, like Ron Dicker, publish articles that cover almost the entire category spectrum. On the flip side, we see about 75% of authors who only write for one specific category. Thus, the author feature could potentially be used as a feature in a prediction model. </p>

<p> In an effort to dive a bit deeper into the author feature, it would be interesting to look at some visualizations of the top authors (based on article counts). The distribution of this relationship could help inform a model (especially if the distribution is long-tailed). </p>

## Section 2: NLP Analysis

### Word2Vec Model

<p> This model is useful for visualizing the semantic similarity between words in our corpus. Since the corpus is very large, we could take a few approaches to modeling: </p>
1. Sampling the dataset
2. Analyzing by category
<p>For the Word2Vec visualization, we'll take the approach of analyzing by category, specifically with the politics category. </p>

```python
from gensim.models import Word2Vec
from sklearn.decomposition import PCA

# training data
politics_df = preproc_df.loc[preproc_df['category'] == 'POLITICS']
sentences = politics_df['tokenized_text'].values

# model training
model = Word2Vec(sentences, min_count=1)

# fit PCA model to the vectors
X = model[model.wv.vocab]
pca = PCA(n_components=2)
result = pca.fit_transform(X)

# create scatter plot of the projection
figure, ax = plt.subplots(1, 1, figsize = (15, 5))
ax.scatter(result[:, 0], result[:, 1])
words = list(model.wv.vocab)
for i, word in enumerate(words):
    ax.annotate(word, xy=(result[i, 0], result[i, 1])) 
plt.show()
```

<p> The Word2Vec algo places words together that appear in similar contexts. This "similarity" is measured by cosine distance (i.e. the dot product of the word vectors). The above graph shows most words are clustered very tightly together, with others interspersed near the cluster. Given a random word, we can look at its most similar words: </p>

```python
model.most_similar('tax', topn=5)
```

```python
model.most_similar('child', topn=5)
```

<p> We can also use the model to measure the similarity of two specific words. The closer to 1 the similarity measure is, the more the two words appear in similar contexts. For example: </p>

```python
model.similarity(w1 = 'family', w2 = 'home')
```

## Section 3: Baseline models for predicting categories 

<p> In this section, we can develop some simple baselines models for predicting categories. We use a typical 65/35 training test split (we could be more robust later by using k-fold cross validation). The features used are the author field, and the aggregated headline + short descriptions field. Based on the performance of these models, we could potentially do some model stacking to develop a more robust model and improve out-of-sample prediction accuracy. A models predictions is only as good as its feature inputs, so another approach would be to spend some serious time on feature engineering. </p>

```python
preproc_df['concat_text'] = preproc_df['authors'] + " " + preproc_df['text'] 
```

### SGD Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score

x = preproc_df['concat_text']
y = preproc_df['category']

## create test and training data
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.35, random_state=21)
```

```python
print("xtrain shape:", xtrain.shape)
print("xtest shape:", xtest.shape)
print("ytrain shape:", ytrain.shape)
print("ytest shape:", ytest.shape)
```

<p> We can start off by creating a tf-idf matrix of our corpus, to use as training data:  </p>

```python
word_vecs = TfidfVectorizer(sublinear_tf=True,analyzer='word', ngram_range=(1, 10))
word_vecs.fit(xtrain)
```

```python
X = word_vecs.transform(xtrain)
```

```python
sgd_cls = SGDClassifier(max_iter=5, learning_rate='optimal')
sgd_cls.fit(X, ytrain)
```

```python
predictions = sgd_cls.predict(word_vecs.transform(xtest))
accuracy = accuracy_score(ytest, predictions)
print("SGD Classifier Test set accuracy: ", accuracy)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(random_state = 31)
log_reg.fit(X, ytrain) # use tf-idf vector computed for SGD Classifer 
log_reg_predictions = log_reg.predict(word_vecs.transform(xtest))
log_reg_accuracy = accuracy_score(ytest, log_reg_predictions)
print("Logistic Regression Test set accuracy: ", log_reg_accuracy)
```

### Linear SVM

```python
from sklearn.svm import LinearSVC

lin_svc = LinearSVC(random_state = 41)
lin_svc.fit(X, ytrain)
lin_svc_predictions = lin_svc.predict(word_vecs.transform(xtest))
lin_svc_accuracy = accuracy_score(ytest, lin_svc_predictions)
print("SVM Test set accuracy: ", lin_svc_accuracy)
```

<p> The SGD Classifer, Logistic Regression model, and Linear SVM model, provide us with solid baseline predictions and achieve a decent level of test set accuracy (~70%). If we wanted to experiment further, we could then use these models to do ensembling and combine the predictions, and this would likely improve prediction accuracy and our model's robustness. As previously mentioned, we could also spend more time on feature engineering and perform cross validation to further improve the model. </p>
